```python
import pandas as pd
import numpy as np
```

```python
hsv_one_df = pd.read_json('hsv_one_df.json')
```

```python
hsv_one_df = hsv_one_df.reindex(range(0, len(hsv_one_df)))
```

```python
hsv_one_df['LBXHE1'] = hsv_one_df['LBXHE1'].where(hsv_one_df['LBXHE1'] != 2, 0)
```

```python
indices = np.random.choice(hsv_one_df[hsv_one_df['LBXHE1']==1].index,
                           int(0.8*len(hsv_one_df[hsv_one_df['LBXHE1']==1])), replace = False) 
flu = np.zeros(hsv_one_df.shape[0], dtype=int)
```

```python
for index in indices:
    flu[index] = 1
hsv_one_df['Flu'] = flu
hsv_one_df = hsv_one_df.set_index('SEQN')
hsv_one_df.head()
```

```python
hsv_one_df.to_json('hsv_one_finale.json')
```
